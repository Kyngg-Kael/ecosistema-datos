```python
import geopandas as gpd
import pandas as pd
from pathlib import Path
import os
```

```python
project_root = Path(os.getcwd()).parent
gpkg_path = project_root / 'data' / 'raw' / 'datos.gpkg'
```

```python
# Revisar columnas
try:
    print("ðŸ” Inspeccionando columnas de 'frontera_agricola_jun2025'...")
    gdf = gpd.read_file(gpkg_path, layer='frontera_agricola_jun2025', rows=1)
    
    print("\nNOMBRE DE COLUMNAS ENCONTRADAS:")
    print(gdf.columns.tolist())
    
    print("\nEJEMPLO DE DATOS:")
    print(gdf.drop(columns='geometry').iloc[0])
    
except Exception as e:
    print(f"Error: {e}")
```

```python
# Capas para el diagnÃ³stico
capas_a_analizar = [
    'runap__registro_unico_nacional_ap',
    'consejos_comunitarios',
    'centro_poblado',
    'zonas_de_reserva_campesina',
    'ley_70_1993'
]
```

```python
print(f"ðŸ“‚ Analizando capas faltantes...\n")

for capa in capas_a_analizar:
    try:
        # Leemos solo 1 fila para ver estructura y nombres
        gdf = gpd.read_file(gpkg_path, layer=capa, rows=1)
        
        # Filtramos columnas de geometrÃ­a para que sea legible
        cols = gdf.columns.tolist()
        if 'geometry' in cols: cols.remove('geometry')
        
        print(f"ðŸ”µ CAPA: {capa}")
        print(f"   Columnas disponibles: {cols}")
        
        # Mostramos el primer dato para ver cuÃ¡l parece ser el nombre
        primer_dato = gdf.iloc[0].drop('geometry', errors='ignore').to_dict()
        print(f"   Dato ejemplo: {primer_dato}")
        print("-" * 50)
        
    except Exception as e:
        print(f"âŒ Error o capa vacÃ­a/inexistente: {capa} -> {e}")
```
